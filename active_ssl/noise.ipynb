```python
# Import
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
from scipy.linalg import eigh, pinvh

from rates.auxillary import (
    label_graph,
    cat2one_hot,
)
from rates.config import SAVE_DIR
from rates.datasets import circles, mixture
from rates.architectures import exp_kernel

# Config
np.random.seed(0)
plt.rc('text', usetex=True)
plt.rc('text.latex', preamble=r'\usepackage{mathtools}')
plt.rc('font', size=10, family='serif', serif='cm')

noise_level = 0
# circles = mixture
# noise_level = .3
```

```python
sigma = .2
lambd = 1e-3
```

```python
n_train = 300
d = 4
n_test = 1000

x_test, y_test = circles(n_test, d=d, noise_level=noise_level, labels=True, test=True)
one_hot = cat2one_hot(y_test.astype(int) - 1)
```

```python
# np.random.seed(0)

nb_trials = 100
nb_trials_distill = 1
nb_distill = (d * (n_train // d)** 2 - n_train) // 2
mul_factor = 10
nb_distill = int(n_train / mul_factor)
errors = np.zeros((nb_trials, nb_distill))

for i in range(nb_trials):
    x, y = circles(n_train, d=d, noise_level=noise_level, labels=True, test=False)

    ind = list(np.random.permutation(np.arange(n_train)))
    y_noise = np.random.randint(1, 5, n_train)

    K = exp_kernel(x, x, sigma=sigma)
    K_test = exp_kernel(x_test, x, sigma=sigma)
    K_inv = pinvh(K, atol=1e-5 / n_train)
    K_inv *= lambd

    # Precomputation
    tmp_x = K @ K_inv
    tmp_x_test = K_test @ K_inv

    for j in range(nb_distill):
        for t in range(mul_factor):
            ind_t = ind.pop()
            y[ind_t] = y_noise[ind_t]

        T = 2 * label_graph(y) - 1

        T_l = T - K_inv
        w, v = eigh(T_l, subset_by_index=[len(T_l) - 5, len(T_l) - 1])

        # ...compute best test error
        phi = tmp_x_test @ v
        w = np.linalg.solve(phi.T @ phi, phi.T @ one_hot)
        pred = np.argmax(phi @ w, axis=1)
        errors[i, j] = (pred + 1 != y_test).mean()

    print(i, end=",", flush=True)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(3, 1.5))
tmp_x = mul_factor * np.arange(nb_distill)

ax.plot(tmp_x, errors.mean(axis=0), label="contrastive")
ax.fill_between(tmp_x, errors.mean(axis=0) - errors.std(axis=0), errors.mean(axis=0) + errors.std(axis=0), alpha=.25)

ax.set_xlabel("Number of corrupted labels")
ax.set_ylabel("Test error")
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(8)
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(8)
# ax.legend(fontsize=8, loc="upper right")
ax.grid(alpha=.5)
# fig.savefig(SAVE_DIR / "noise.pdf", bbox_inches='tight', pad_inches=0)
```
